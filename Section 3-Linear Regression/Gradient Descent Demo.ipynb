# Gradient Descent Demo

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=(x-5)**2+3
print(x,y)
```

```python
plt.style.use("seaborn")
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

```

```python
# gradient descent
#initialize x with any random value
x=0
lr=0.1 #learning rate
plt.plot(x,y)
for i in range(50):
    gradient=2*(x-5) #derivate of the function gradient
    x=x-lr*gradient #gradient
    y=(x-5)**2+3
    plt.scatter(x,y)
    print(x)
```
